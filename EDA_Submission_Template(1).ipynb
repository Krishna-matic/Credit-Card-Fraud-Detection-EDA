## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1 visualization code
# Set the plot style
sns.set_theme(style="whitegrid", context="talk")
# Create the figure
plt.figure(figsize=(8, 5))
# Count Plot
ax = sns.countplot(
    data=df,
    x='Class',
    hue='Class',
    palette='Set2',
    legend=False
)
# Title and Labels
plt.title(
    "Distribution of Fraudulent and Legitimate Transactions",
    fontsize=16,
    fontweight='bold'
)
plt.xlabel("Transaction Type", fontsize=12)
plt.ylabel("Number of Transactions", fontsize=12)
# Rename x-axis labels
plt.xticks([0, 1], ["Legitimate", "Fraud"])
# Add value labels on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt='{:,.0f}', fontsize=11)
# Adjust layout
plt.tight_layout()
# Display the plot
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer Here.

A count plot was chosen because the target variable (Class) is categorical with two categories: Legitimate and Fraudulent transactions. It provides a clear comparison of the frequency of each class and helps identify any class imbalance in the dataset.

##### 2. What is/are the insight(s) found from the chart?

Answer Here


*   The dataset contains significantly more legitimate transactions than fraudulent ones.


*   After removing duplicate records, there are 284,315 legitimate transactions and 492 fraudulent transactions.
*   The chart clearly indicates that the dataset is highly imbalanced, with fraud representing only a very small fraction of all transactions.





##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.


Answer Here

The severe class imbalance highlights one of the biggest challenges in fraud detection. A model trained on such data may become biased toward predicting legitimate transactions. Recognizing this imbalance early helps in selecting appropriate techniques such as resampling, class weighting, or anomaly detection to improve fraud detection performance.

#### Chart - 3

````python
# Chart - 3 visualization code
plt.figure(figsize=(8,5))
sns.histplot(
    data=df,
    x='Amount',
    bins=60,
    kde=True,
    color='#4C72B0'
)
plt.xscale('log')
plt.title("Distribution of Transaction Amounts (Log Scale)",
          fontsize=16,
          fontweight='bold')

plt.xlabel("Transaction Amount (Log Scale)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer Here.

A histogram was chosen because the Amount feature is a continuous numerical variable, and histograms are effective for understanding its distribution. Since transaction amounts vary over a very wide range and are highly right-skewed, a logarithmic scale was applied to the x-axis. This improves readability by spreading out smaller transaction values while still displaying larger transactions, making the overall distribution easier to interpret.

##### 2. What is/are the insight(s) found from the chart?

Answer Here


*   The distribution of transaction amounts is highly right-skewed, indicating that most transactions involve relatively small amounts.
*   A small number of transactions have significantly larger values, creating a long right tail in the distribution.


*   Applying a logarithmic scale reveals the spread of smaller transaction amounts more clearly, which would otherwise be compressed on a standard linear scale.

*   The presence of high-value transactions suggests potential outliers that should be investigated further during the analysis.





##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.


Answer Here

Understanding the distribution of transaction amounts helps financial institutions establish normal customer spending patterns. Since most transactions are of lower value, unusually large transactions can be flagged for additional verification as part of a fraud detection strategy. However, transaction amount alone cannot determine whether a transaction is fraudulent because legitimate customers may also make high-value purchases. Therefore, transaction amount should be analyzed together with other features to improve fraud detection accuracy and reduce false positives.

#### Chart - 8

````python
# Correlation with target variable
corr = df.corr()['Class'].drop('Class')
# Select top 10 strongest correlations (absolute values)
top_corr = corr.loc[corr.abs().sort_values(ascending=False).head(10).index]
top_corr = top_corr.sort_values()
# Set theme
sns.set_theme(style="whitegrid", context="talk")
# Create figure
plt.figure(figsize=(9,6))
# Horizontal Bar Chart
ax = sns.barplot(
    x=top_corr.values,
    y=top_corr.index,
    hue=top_corr.index,
    palette="viridis",
    legend=False
)
# Add correlation values
for i, value in enumerate(top_corr.values):
    ax.text(
        value,
        i,
        f' {value:.3f}',
        va='center',
        fontsize=10
    )
plt.title(
    "Top 10 Features Most Correlated with Fraud",
    fontsize=16,
    fontweight='bold'
)
plt.xlabel("Correlation Coefficient")
plt.ylabel("Features")
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer Here.

A horizontal bar chart was chosen because it clearly ranks the features according to the strength of their correlation with the target variable (Class). Displaying both positive and negative correlations helps identify the variables that are most strongly associated with fraudulent transactions, making feature importance easier to interpret.

##### 2. What is/are the insight(s) found from the chart?

Answer Here



*   V17, V14, and V12 exhibit the strongest negative correlations with fraudulent transactions, indicating they are highly informative features for distinguishing fraud.

*   V11, V4, and V2 show the strongest positive correlations with the target variable.
*   Most features have relatively weak correlations with fraud, suggesting that fraud detection depends on the combined influence of multiple variables rather than a single feature


*   The presence of both positive and negative correlations highlights the importance of considering feature interactions when building a fraud detection model.

Answer Here
