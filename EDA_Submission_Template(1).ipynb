# **Project Summary -**

This project focuses on the Exploratory Data Analysis (EDA) of a credit card fraud detection dataset to understand the characteristics of fraudulent and legitimate transactions and identify meaningful patterns that can support the development of an effective fraud detection system. Credit card fraud is a significant challenge for financial institutions as it leads to substantial financial losses and negatively impacts customer trust. The primary objective of this project was to analyze the dataset, discover important trends, identify influential features, and provide business recommendations that can improve fraud detection.

The dataset contains 284,807 credit card transactions with 31 variables, including Time, Amount, Class, and 28 anonymized features (V1–V28) generated using Principal Component Analysis (PCA) to protect sensitive customer information. The target variable, Class, indicates whether a transaction is legitimate (0) or fraudulent (1). Initial data exploration revealed that the dataset contains no missing values, ensuring high data quality. Duplicate records were identified and analyzed before proceeding with the visualization and exploratory analysis.

Several statistical summaries and visualizations were performed to understand the structure and characteristics of the data. The analysis showed that the dataset is highly imbalanced, with only 492 fraudulent transactions compared to 284,315 legitimate transactions. This imbalance highlights one of the major challenges in fraud detection, as predictive models trained on such datasets require specialized techniques to avoid bias toward the majority class.

The exploratory analysis included various visualizations such as class distribution, transaction amount distribution, transaction time analysis, box plots, violin plots, scatter plots, correlation analysis, pair plots, and heatmaps. These visualizations helped uncover important relationships between variables and provided valuable insights into fraud patterns. The transaction amount analysis indicated that fraudulent transactions frequently occur in lower transaction amount ranges, suggesting that fraud monitoring should not focus exclusively on high-value transactions. Time-based analysis showed fluctuations in transaction activity throughout the observation period, while scatter plots and pair plots revealed noticeable separation between fraudulent and legitimate transactions when multiple important features were considered together.

Correlation analysis identified V17, V14, V12, V10, and V11 as the features most strongly associated with fraudulent transactions. Further investigation using box plots and pair plots confirmed that these variables exhibit clear differences between legitimate and fraudulent transactions. The correlation heatmap also demonstrated that most PCA-transformed features have low correlations with one another, indicating minimal multicollinearity and suggesting that each feature contributes unique information to the dataset. These findings emphasize that fraud detection cannot rely on a single feature and instead requires combining multiple variables for accurate classification.

Based on the insights obtained from this exploratory analysis, it is recommended that financial institutions implement machine learning-based fraud detection systems capable of analyzing multiple transaction characteristics simultaneously. Special attention should be given to the most influential features identified during the analysis, while also monitoring low-value transactions that may otherwise receive less scrutiny. By leveraging these findings, organizations can improve fraud detection accuracy, reduce financial losses, minimize false positives, and strengthen customer confidence in digital payment systems.

Overall, this project demonstrates how Exploratory Data Analysis serves as a critical first step in understanding complex financial datasets. The insights generated through this analysis provide a strong foundation for developing robust fraud detection models and enable data-driven decision-making to combat fraudulent financial activities effectively.

#### Chart - 1

````python
# Chart - 1 visualization code
# Set the plot style
sns.set_theme(style="whitegrid", context="talk")
# Create the figure
plt.figure(figsize=(8, 5))
# Count Plot
ax = sns.countplot(
    data=df,
    x='Class',
    hue='Class',
    palette='Set2',
    legend=False
)
# Title and Labels
plt.title(
    "Distribution of Fraudulent and Legitimate Transactions",
    fontsize=16,
    fontweight='bold'
)
plt.xlabel("Transaction Type", fontsize=12)
plt.ylabel("Number of Transactions", fontsize=12)
# Rename x-axis labels
plt.xticks([0, 1], ["Legitimate", "Fraud"])
# Add value labels on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt='{:,.0f}', fontsize=11)
# Adjust layout
plt.tight_layout()
# Display the plot
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer Here.

#### Chart - 8

````python
# Correlation with target variable
corr = df.corr()['Class'].drop('Class')
# Select top 10 strongest correlations (absolute values)
top_corr = corr.loc[corr.abs().sort_values(ascending=False).head(10).index]
top_corr = top_corr.sort_values()
# Set theme
sns.set_theme(style="whitegrid", context="talk")
# Create figure
plt.figure(figsize=(9,6))
# Horizontal Bar Chart
ax = sns.barplot(
    x=top_corr.values,
    y=top_corr.index,
    hue=top_corr.index,
    palette="viridis",
    legend=False
)
# Add correlation values
for i, value in enumerate(top_corr.values):
    ax.text(
        value,
        i,
        f' {value:.3f}',
        va='center',
        fontsize=10
    )
plt.title(
    "Top 10 Features Most Correlated with Fraud",
    fontsize=16,
    fontweight='bold'
)
plt.xlabel("Correlation Coefficient")
plt.ylabel("Features")
plt.tight_layout()
plt.show()
````
